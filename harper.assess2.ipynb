```python
# 2. Rather than deal with these missing features explicitly, we will remove any afflicted examples for now. 
# Please filter any examples with missing features, using both “grep” and using “python”.
import json

count = 0
data = []

with open('ad.json') as f:
    for line in f:
        if '?' not in line:
            data.append(json.loads(line))
        else: count +=1
print "There are " + str(count) + " lines with missing features."
        
```
